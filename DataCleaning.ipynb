```python
crime = []
count = 0

for aa in tmpMain:
    a = re.findall('(?<=犯|因|[共同])([\u4E00-\u9FFF]*[罪|人])[?>=，]', aa)
    if a != []:
        crime.append(a)
    else:
        if "無罪" in aa:
            crime.append('無罪')
        else:
            crime.append(aa)
# print(crime)
```

```python
crimeLength = []
for data in crime:
    if 2 < len(data) <= 10:
        if '上訴駁回' in data:
            crimeLength.append('NotAbout罪類')
        else:
            crimeLength.append('數罪併罰')
    elif len(data) == 1:
        for i in data:
            if '傷害直系血親尊親屬致死' in i:
                crimeLength.append('傷害直系血親尊親屬致死') 
            elif '傷害直系血親尊親屬' in i:
                crimeLength.append('傷害直系血親尊親屬')
            elif '傷害致死' in i:
                crimeLength.append('傷害致死')
            elif '過失傷害致重傷' in i:
                crimeLength.append('過失傷害致重傷')
            elif '傷害致重傷' in i:
                crimeLength.append('傷害致重傷')
            elif '重傷害未遂' in i:
                crimeLength.append('重傷害未遂')
            elif '重傷致死' in i:
                crimeLength.append('重傷致死')
            elif '重傷害' in i:
                crimeLength.append('重傷害')
            elif '教唆傷害' in i:
                crimeLength.append('教唆傷害')    
            elif '逃逸' in i:
                crimeLength.append('肇事逃逸')    
            elif '過失傷害' in i:
                crimeLength.append('過失傷害')
            elif '傷害' in i:
                crimeLength.append('傷害')
            else:
                crimeLength.append('NotAbout傷害')
    else:
        if '無罪' in data:
            crimeLength.append(data)
        else:    
            crimeLength.append('NotAbout罪類')
```

```python
data = pd.read_csv('D:/期末專題/AI法官/rawData/forModel.csv', encoding='utf-8')
data_main = data.loc[:,'judge_main']
data_content = data.loc[:, 'judge_content']
# data_main
data_content
```

```python
name_main = []

r_name_main = '[:|：]([\u4E00-\u9FFF]{3})'
for main in data_main:
    try:
        nm = re.search(r_name_main, main).group(1)
        name_main.append(nm)
    except:
        name_main.append(main)

name_main1 = []
# 篩掉 '原判決' 與 未適配名字
for a in name_main:
    if len(a) == 3:
        if a != '原判決':
            name_main1.append(a)
    
# name_main1



name_content = []
test = 0
r_name_content = '[證人]([\u4E00-\u9FFF]{3})[於]'
for content in data_content:
    try:
        nc = re.search(r_name_content, content).group(1)
        name_content.append(nc)
    except:
        test+=1

        

        
stop = ['均不致', 
        '及被告', 
        '聞聲後', 
        '當時對', 
        '據非出', 
        '並立即', 
        '並認為', 
        '明有出', 
        '明被告', 
        '據能力', 
        '前揭關',
        '明係出',
        '之父對',
        '與被告',
        '之車輛',
        '之左足',
        '代表人',
        '當時對',
        '人未能',
        '之母對',
        '返家後',
        '被告出',
        '在店內',
        '均不至',
        '據亦未',
        '據資料',
        '人溫滿',
        '既然有',
        '雖曾住',
        '詢問關',
        '固承認',
        '人嚴竣',
        '有同意',
        '攻擊出',
        '則受僱',
        '人未曾',
        '應亦有',
        '機車過'
        ]
name_content1 = []
# 篩掉 '原判決' 與 未適配名字
for a in name_content:
    if len(a) == 3:
        for s in stop:
            if a != s:
                name_content1.append(a)
name_content2 = list(set(name_content1))
name_content2

for m in name_main1:
    name_content2.append(m)
# name_content2
    
with open ('./name_stop.txt', 'w', encoding='utf-8') as f:
    for n in name_content2:
        nn = ''.join(n)+'\n'
        f.write(nn)
```

`篩出 檢察官/法官 人名`

```python
r_name_prosecutor = '(?<=檢察官)([\u4E00-\u9FFF]{3})[到|提]'
r_name_judge = '(?<=庭法官)([\u4E00-\u9FFF]{3})'

name_pros_judg = []
tmpc = 0
for content in data_content:
    try:
        pros = re.search(r_name_prosecutor, content).group(1)
        name_pros_judg.append(pros)
        judg = re.search(r_name_judge, content).group(1)
        name_pros_judg.append(judg)
    except:
        tmpc+=1
        
name_pros_judg_ = list(set(name_pros_judg))
        
with open ('./name_pros_judg.txt', 'w', encoding='utf-8') as f:
    for n in name_pros_judg_:
        nn = ''.join(n)+'\n'
        f.write(nn)
```
