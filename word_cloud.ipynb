```python
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
# 讀取每首歌的前10個tags
text = open("D:/期末專題/AI法官/rawData/crimeCategory/crimeCategory_1_傷害直系血親尊親屬致死_jeba.txt",encoding="utf-8").read()
# 設定停用字(排除常用詞、無法代表特殊意義的字詞)
# stopwords = {}.fromkeys(["沒有","一個","什麼","那個",'令狐','甚麼','田伯光','林平之','儀琳','師父','說道','定逸','林震南'])
```

```python
# the font from github: https://github.com/adobe-fonts

crime = ['傷害直系血親尊親屬致死', 
         '傷害直系血親尊親屬', 
         '傷害致死', 
         '過失傷害致重傷', 
         '傷害致重傷', 
         '重傷害未遂', 
         '重傷致死', 
         '重傷害',
         '教唆傷害', # 其他
         '肇事逃逸', # 其他
         '過失傷害',
         '傷害',
         '無罪']




import imageio
back_color = imageio.v2.imread('D:/期末專題/AI法官/gavel1.png')


font = r'C:/Windows/Fonts/msjh.ttc'
wc = WordCloud(collocations=False,
               width=800,
               height=800,
               font_path=font, 
               margin=2, 
               mask=back_color, 
               background_color='white',
               contour_width=0.5,
               contour_color='black',
               max_words = 250).generate(text)


plt.imshow(wc)
plt.axis("off")
plt.show()

# wc.to_file('gavel_傷害直系血親尊親屬致死_border.png')  # 把詞雲保存下来 
```

```python
# 產生文字雲
wc = WordCloud(font_path="C:/Windows/Fonts/NotoSansTC-Regular.otf", #設置字體
               background_color="white", #背景顏色
               max_words = 2000 )        #文字雲顯示最大詞數
#                stopwords=stopwords)      #停用字詞
wc.generate(text)
# 視覺化呈現
plt.imshow(wc)
plt.axis("off")
plt.figure(figsize=(10,6), dpi = 100)
plt.show()
```
