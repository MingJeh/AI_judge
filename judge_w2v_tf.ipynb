```python
import numpy as np
import pandas as pd
```

```python
from gensim.models import word2vec
# test model

# load model
model = word2vec.Word2Vec.load('word2vec_judge_sg_w10_v300_e15.model')
```

`word2vec`

```python
key = [
       '傷害直系血親尊親屬致死', 
       '傷害直系血親尊親屬', 
       '傷害致死', 
       '過失傷害致重傷', 
       '傷害致重傷', 
       '重傷害未遂', 
       '重傷致死', 
       '重傷害',
       '教唆傷害',
       '逃逸',
       '過失傷害',
       '傷害'
]
# print(key)


# print(model.wv[key].shape)
w2v_simi_lst = []
for item in model.wv.most_similar(key[9], topn=200):
    w2v_simi_lst.append(item[0])
#     print(type(item)) -> tuple
print(w2v_simi_lst)

# 計算相似程度
# model.wv.similarity('傷害致死', '傷害直系血親尊親屬致死')
```

`tf`

```python

folderPath_name = [
         '1_傷害直系血親尊親屬致死', 
         '2_傷害直系血親尊親屬', 
         '3_傷害致死', 
         '4_過失傷害致重傷', 
         '5_傷害致重傷', 
         '6_重傷害未遂', 
         '7_重傷致死', 
         '8_重傷害',
         '9_教唆傷害',
         '10_肇事逃逸',
         '11_過失傷害',
         '12_傷害',
         '13_無罪']


folderPath = f'D:/期末專題/AI法官/rawData/crimeCategory/TF-IDF/{folderPath_name[9]}.csv'
tf_data = pd.read_csv(folderPath, encoding='utf-8')
tf_word = tf_data.iloc[:,0:2]
c_lst = []
for col in tf_data.columns:
    c_lst.append(col)
top = c_lst[1]  # 取得 tf 總數的 column
word_top = c_lst[0]  # 取得 tf word的 column
tf_word_sort = tf_word.sort_values(by=[top], ascending=False)
tf_word_sort_top50 = tf_word_sort.iloc[0:200,:]  # -> dataframe
tf_word_lst = tf_word_sort_top50[word_top].values[:].tolist()
tf_word_lst  # 取得 tf word
```

`找出 w2v 和 tf 都出現的 word`

```python
# w2v_simi_lst
# tf_word_lst
w2v_tf_lst = []  # 紀錄 w2v 和 tf 都出現的 word

for ww in w2v_simi_lst:
    for wf in tf_word_lst:
        if ww == wf:
            w2v_tf_lst.append(ww)
    
while len(w2v_tf_lst) < 200:
    w2v_tf_lst.append(' ')
    
            
print(w2v_tf_lst)
```

```python
crime = [
         '1_傷害直系血親尊親屬致死', 
         '2_傷害直系血親尊親屬', 
         '3_傷害致死', 
         '4_過失傷害致重傷', 
         '5_傷害致重傷', 
         '6_重傷害未遂', 
         '7_重傷致死', 
         '8_重傷害',
         '9_教唆傷害',
         '10_肇事逃逸',
         '11_過失傷害',
         '12_傷害',
         '13_無罪'
]
# w2v_simi_lst
# tf_word_lst
# w2v_tf_lst

origin = {
    'w2v': w2v_simi_lst,
    'tf': tf_word_lst
}

total = {
    'w2v&tf': w2v_tf_lst
}

df1 = pd.DataFrame(origin)
df1.insert(2, 'w2v&tf', w2v_tf_lst)



folderPath_w2v_tf = 'D:/期末專題/AI法官/rawData/crimeCategory/TF-IDF/'
df1.to_csv(f'{folderPath_w2v_tf}/w2v_tf_top50_{crime[9]}.csv', encoding='utf-8-sig', index=False)
```
